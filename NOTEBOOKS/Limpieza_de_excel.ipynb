```python
cleaned_dfs = {}

for year, df_year in dfs.items():
    pair1 = df_year.iloc[:, [0, 1]].copy()
    pair1.columns = ["cuenta", "valor"]

    pair2 = df_year.iloc[:, [3, 4]].copy()
    pair2.columns = ["cuenta", "valor"]

    combined = pd.concat([pair1, pair2], ignore_index=True)
    combined = combined.dropna(subset=["cuenta"])
    cleaned_dfs[year] = combined

for year, df_clean in cleaned_dfs.items():
    print(year)
    display(df_clean)
```

```python
for year, df_year in cleaned_dfs.items():
    cleaned_dfs[year] = df_year.rename(columns={"valor": year[3:]+"_valor"})

dfs_list = list(cleaned_dfs.values())
cleaned_merged_df = dfs_list[0]
for df_year in dfs_list[1:]:
    cleaned_merged_df = pd.merge(cleaned_merged_df, df_year, on="cuenta", how="outer", sort=False)

display(cleaned_merged_df)
```
