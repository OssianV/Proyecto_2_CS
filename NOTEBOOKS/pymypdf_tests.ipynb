# Tests using PyMuPDF and OpenCV


To access a supported document (in this case a .pdf) we need to use the following command:

```python
import sys
import cv2 as cv
from pathlib import Path
import pymupdf
import matplotlib.pyplot as plt
doc = pymupdf.open("../ATTACHMENTS/_pdf/Estados-Financieros-MetLife-Mexico-SA-de-CV-2025.pdf")
```

This created the `document` object `doc`. Any document contains many metainformation (metadata), like the author, number of pages, etc. For now this metadata wont be of our interest.

Since the pdf always contains only one page, we need to store said page into a `Page` object page. We do that next:

```python
page = doc[0]    # Type: pymupdf.Page
page_raster_image = page.get_pixmap(dpi=500)    # Type: pymupdf.Pixmap.

# We specify the dpi to get a better quality image, the default is 72 dpi,
# which is very low for our purposes. The higher the dpi, the better the
# quality of the image, but also the larger the file size.
```

Observe we also defined a `Pixmap` object called `page_raster_image`. This object stores an RBG imgage of the page.

We now save the page image in a file using:

```python
# To erase the image created we use the pathlib library
file_path = Path(f"../ATTACHMENTS/_png/page_{page.number}.png")
file_path.unlink(missing_ok=True)

page_raster_image.save(file_path)
```

The resulting png look like this:

![imagen](../ATTACHMENTS/_png/page_0.png)

From the png we can easily see that a common pattern that could help us develop a process for the correct parsing of the information contained in the financial statements, is that both finantial statements start just bellow a big horizontal black-colored rectangle that contains the words **MetLife**.

Based on this finding, we can plan for a script that would do two things:

1. Search and identify the clearly visible black rectangle;
2. Apply OCR (Optical Character Recognition) to verify that said rectangle contains the word "MetLife".

For that we will need to use the library OpenCV (Open Computer Vision).


```python
img = cv.imread(cv.samples.findFile(str(file_path)), cv.IMREAD_GRAYSCALE)

output_image_path = Path(f"../OUTPUTS/page_{page.number}_cut.png")

if img is not None:
    output_image_path.unlink(missing_ok=True)
    cv.imwrite(str(output_image_path), img)
else:
    sys.exit("Could not read the image.")

# if img is None:
#     sys.exit("Could not read the image.")

# cv.imshow("Display window", img)
# k = cv.waitKey(0)    # We define that the window will wait indefinitely (value 0) for user input to close it.
# if k == ord("s"):    # We store the key pressed in the variable k, and if it is the "s" key, we save the image.
#     output_image_path.unlink(missing_ok=True)
#     cv.imwrite(str(output_image_path), img)
```
