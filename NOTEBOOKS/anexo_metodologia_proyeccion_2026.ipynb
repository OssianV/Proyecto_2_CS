## Estructura de la proyeccion y notacion utilizada

La proyeccion se construye siguiendo la estructura del estado de resultados y respetando las identidades contables entre cuentas. Primero se proyecta el total anual de 2026; solo despues se distribuye mes a mes.

La notacion utilizada en este anexo es la siguiente:

- $PE_t$: primas emitidas en el ano $t$.
- $PC_t$: primas cedidas en el ano $t$.
- $PR_t$: primas retenidas en el ano $t$.
- $\Delta RRRC_t$: incremento neto en la reserva de riesgos en curso y fianzas en vigor en el ano $t$.
- $PRD_t$: primas retenidas devengadas en el ano $t$.
- $CAdq_t$: costo neto de adquisicion en el ano $t$.
- $CNS_t$: costo neto de siniestralidad en el ano $t$.
- $UBr_t$: utilidad o perdida bruta en el ano $t$.
- $COp_t$: costo neto de operacion en el ano $t$.
- $UOp_t$: utilidad o perdida de la operacion en el ano $t$.
- $RI_t$: rubro "Resultado de inversiones" dentro del RIF en el ano $t$.
- $RIF_t$: resultado integral del financiamiento en el ano $t$.
- $UAI_t$: utilidad antes de ISR en el ano $t$.
- $ISR_t$: provision para el ISR en el ano $t$.
- $UN_t$: utilidad neta del ejercicio en el ano $t$.

Las identidades contables que guian el modelo son:

$$
PR_t = PE_t - PC_t
$$

$$
PRD_t = PR_t - \Delta RRRC_t
$$

$$
UBr_t = PRD_t - CAdq_t - CNS_t
$$

$$
UOp_t = UBr_t - COp_t
$$

$$
UAI_t = UOp_t + RIF_t
$$

$$
UN_t = UAI_t - ISR_t
$$

Dado que solo se cuenta con informacion anual consolidada, el modelo se plantea como una proyeccion por escenarios basada en razones historicas y no como una proyeccion actuarial detallada a nivel poliza.


## Proyeccion de ingresos

En esta seccion se proyectan las cuentas que determinan el ingreso tecnico del ejercicio antes de gastos y de avaluo financiero. El criterio general es proyectar primero el total anual de 2026 y despues construir el desglose mensual.

### Prima emitida

El criterio inicialmente sugerido fue utilizar el promedio interanual de las tasas de crecimiento de las primas emitidas. Sin embargo, la serie 2021-2025 presenta un crecimiento inusualmente alto en 2022 y 2023, por lo que usar un unico promedio simple sobre todo el periodo puede sobreestimar la tendencia reciente.

Se define la tasa de crecimiento anual como:

$$
g_t = \frac{PE_t}{PE_{t-1}} - 1, \qquad t \in \{2022,2023,2024,2025\}
$$

Con esta definicion se comparan tres estadisticos:

$$
\bar g = \frac{1}{4}\sum_{t=2022}^{2025} g_t = 9.69\%
$$

$$
\bar g_{rec} = \frac{g_{2024} + g_{2025}}{2} = 6.79\%
$$

$$
g_{CAGR} = \left(\frac{PE_{2025}}{PE_{2021}}\right)^{1/4} - 1 = 7.65\%
$$

El promedio interanual da el mismo peso a todos los anos y, por ello, queda mas expuesto a anos atipicos. El promedio de los ultimos dos anos captura mejor la trayectoria reciente. El CAGR no se interpreta aqui como una tasa de reinversion, sino como una tasa anual equivalente que resume el crecimiento acumulado entre 2021 y 2025.

**[REVISA]** La hipotesis de que 2022 y 2023 reflejan un impulso anormal asociado al ramo vida despues de la pandemia debe respaldarse con una fuente externa o con una nota sectorial si se quiere presentar como afirmacion fuerte y no solo como interpretacion razonable.

Con base en lo anterior, la prima emitida 2026 se proyecta por escenario de la siguiente forma:

- **Escenario de normalizacion post-pandemia**: se usa $\bar g_{rec}$, pues se asume que el impulso extraordinario ya se disipo y que 2026 debe parecerse mas al comportamiento reciente.
- **Escenario de persistencia parcial del impulso post-pandemia**: se usa $g_{CAGR}$, pues permite reconocer parte del crecimiento acumulado del ciclo sin darle el mismo peso a todos los anos.
- **Escenario de continuidad historica integral**: se usa $\bar g$, que equivale al criterio original de promediar los cuatro crecimientos interanuales observados.

La formula general es:

$$
\widehat{PE}_{2026,s} = PE_{2025}\,(1 + \widehat g_s)
$$

donde $s$ denota el escenario considerado.

### Prima cedida

La prima cedida se proyecta como proporcion de la prima emitida, pues representa la parte del negocio transferida a reaseguro. Se define el ratio historico:

$$
c_t = \frac{PC_t}{PE_t}
$$

Al comparar el promedio simple y el promedio ponderado del periodo completo se obtuvo:

- promedio anual: $3.68\%$;
- promedio ponderado del periodo: $3.71\%$.

Dado que ambos valores son practicamente iguales, se adopta el promedio ponderado del periodo por ser ligeramente mas robusto al dar mas peso a los anos con mayor volumen de primas emitidas:

$$
\widehat c = \frac{\sum_{t=2021}^{2025} PC_t}{\sum_{t=2021}^{2025} PE_t}
$$

y, por tanto:

$$
\widehat{PC}_{2026,s} = \widehat c\,\widehat{PE}_{2026,s}
$$

### Prima retenida

La prima retenida se obtiene por identidad contable:

$$
\widehat{PR}_{2026,s} = \widehat{PE}_{2026,s} - \widehat{PC}_{2026,s}
$$

### Primas retenidas devengadas

Las primas retenidas devengadas se obtienen restando a la prima retenida el incremento neto en la reserva de riesgos en curso y fianzas en vigor. En vez de proyectar primero el incremento neto de la reserva, se considera mas limpio proyectar directamente la tasa de devengamiento de la prima retenida.

Se define:

$$
d_t = \frac{PRD_t}{PR_t}
$$

donde $d_t$ representa la proporcion de la prima retenida que efectivamente se devenga en el mismo ejercicio.

Los estadisticos calculados fueron:

- promedio anual del periodo: $94.57\%$;
- promedio anual de los ultimos dos anos: $96.48\%$;
- promedio ponderado del periodo: $94.68\%$.

La tasa reciente es mayor que las otras dos, lo cual es consistente con la desaceleracion de la emision en 2024 y 2025: cuando se emiten menos polizas nuevas, el incremento de la reserva de riesgos en curso tiende a ser menor y, por tanto, una mayor proporcion de la prima retenida queda devengada en el mismo ejercicio.

Se adopta el siguiente criterio por escenario:

- **Escenario de normalizacion post-pandemia**: $\widehat d_s = 96.48\%$.
- **Escenario de persistencia parcial del impulso post-pandemia**: $\widehat d_s = 94.68\%$.
- **Escenario de continuidad historica integral**: $\widehat d_s = 94.57\%$.

**[REVISA]** Si en tu hoja final decidiste usar el promedio ponderado tambien en el escenario historico, ajusta esta seccion para que la narrativa coincida exactamente con el modelo ejecutado.

Finalmente:

$$
\widehat{PRD}_{2026,s} = \widehat d_s\,\widehat{PR}_{2026,s}
$$

y el incremento neto implicito en la reserva queda dado por:

$$
\widehat{\Delta RRRC}_{2026,s} = \widehat{PR}_{2026,s} - \widehat{PRD}_{2026,s}
$$


## Proyeccion de gastos

### Costo neto de adquisicion

El costo neto de adquisicion representa el costo comercial de generar el negocio del ejercicio, especialmente comisiones y compensaciones a agentes. Dado que se espera que esta cuenta se mueva con el volumen de negocio emitido, se modela como proporcion de la prima emitida.

Se define el ratio historico:

$$
a_t = \frac{CAdq_t}{PE_t}
$$

Los valores observados fueron:

- 2021: $0.139705664$;
- 2022: $0.140378117$;
- 2023: $0.172375256$;
- 2024: $0.151338406$;
- 2025: $0.149766818$.

Se comparan dos formas naturales de resumir esta proporcion:

$$
\bar a = \frac{1}{5}\sum_{t=2021}^{2025} a_t = 15.07\%
$$

y

$$
\bar a_w = \frac{\sum_{t=2021}^{2025} CAdq_t}{\sum_{t=2021}^{2025} PE_t} = 15.14\%
$$

Como ambos resultados son casi iguales, se adopta el promedio ponderado del periodo por ser ligeramente mas robusto al dar mayor peso a los anos con mayor volumen de prima emitida. La proyeccion queda entonces como:

$$
\widehat{CAdq}_{2026,s} = \bar a_w\,\widehat{PE}_{2026,s}
$$

#### Alternativa descartada: proporcion marginal

En una version preliminar se considero usar la razon marginal entre cambios anuales:

$$
m_t = \frac{\Delta CAdq_t}{\Delta PE_t}
$$

y resumirla mediante:

$$
\bar m = \frac{\sum_{t=2022}^{2025} \Delta CAdq_t}{\sum_{t=2022}^{2025} \Delta PE_t}
$$

Sin embargo, esta magnitud no mide la participacion promedio del costo de adquisicion sobre la prima emitida, sino la sensibilidad incremental del costo ante cambios en la emision. Como el objetivo del modelo es proyectar el costo total como proporcion de un nivel de negocio y no como respuesta marginal a cambios de un ano a otro, esta alternativa se descarta.

**[REVISA]** Esta justificacion fue agregada para distinguir con claridad entre una razon de nivel y una razon marginal.


### Costo neto de siniestralidad

Para el costo neto de siniestralidad no se considera adecuado proyectar niveles absolutos por tasa de crecimiento. Lo mas natural es modelar la razon de perdida tecnica, es decir, el costo de siniestralidad por cada peso de prima retenida devengada.

Se define:

$$
\lambda_t = \frac{CNS_t}{PRD_t}
$$

El denominador elegido no es la prima emitida ni la prima retenida, sino la prima retenida devengada. La razon es que la siniestralidad se genera sobre riesgo retenido que ya corrio durante el periodo; usar prima retenida sin ajustar por devengamiento incorporaria negocio cuyo riesgo aun no ha transcurrido y tenderia a subestimar la razon de perdida.

Los estadisticos calculados para $\lambda_t$ fueron:

- promedio anual: $70.08\%$;
- promedio anual de los ultimos dos anos: $71.00\%$;
- promedio ponderado del periodo: $76.65\%$.

Dado que el promedio anual y el promedio reciente son practicamente iguales, mientras que el promedio ponderado queda visiblemente por encima, se concluye que no hay evidencia suficiente para usar un ratio distinto por escenario. En consecuencia, se utiliza el promedio anual para todos los escenarios:

$$
\widehat \lambda = 70.08\%
$$

y la proyeccion queda dada por:

$$
\widehat{CNS}_{2026,s} = \widehat \lambda\,\widehat{PRD}_{2026,s}
$$

**[REVISA]** Si en la version final prefieres enfatizar mas el comportamiento reciente, sustituir $70.08\%$ por $71.00\%$ apenas modifica la proyeccion y puede defenderse igual de bien. Lo importante es no mezclar sin justificacion el promedio ponderado alto con el escenario base.

### Utilidad o perdida bruta

La utilidad o perdida bruta se obtiene por identidad contable:

$$
\widehat{UBr}_{2026,s} = \widehat{PRD}_{2026,s} - \widehat{CAdq}_{2026,s} - \widehat{CNS}_{2026,s}
$$

### Costo neto de operacion

El costo neto de operacion se modela como proporcion de la prima emitida, del mismo modo que el costo de adquisicion, pues ambas cuentas se asocian al tamano operativo del negocio. Los estadisticos calculados fueron:

- promedio anual: $6.32\%$;
- promedio ponderado del periodo: $6.28\%$.

Dado que ambos resultados son casi iguales, se utiliza el promedio ponderado del periodo:

$$
\bar o_w = 6.28\%
$$

$$
\widehat{COp}_{2026,s} = \bar o_w\,\widehat{PE}_{2026,s}
$$

### Utilidad o perdida de la operacion

Finalmente, la utilidad o perdida de la operacion se obtiene como:

$$
\widehat{UOp}_{2026,s} = \widehat{UBr}_{2026,s} - \widehat{COp}_{2026,s}
$$


## Avaluo de activos: resultado integral del financiamiento

El tratamiento del RIF requiere una separacion explicita para evitar doble conteo. El motivo es que, en la proyeccion mensual 2026, el criterio indicado es enviar a reservas el flujo neto mensual y colocar el $50\%$ de dichas reservas en CETES a 28 dias. Si ademas se proyectara el RIF total con una tasa historica global, una parte del rendimiento de esas nuevas reservas ya quedaria incorporada dos veces.

Por ello, el RIF 2026 se descompone en tres partes:

$$
\widehat{RIF}_{2026,s} = \widehat{RI}^{base}_{2026,s} + \widehat{RI}^{cetes}_{2026,s} + \widehat{RIF}^{res}_{2026,s}
$$

donde:

- $\widehat{RI}^{base}_{2026,s}$ es el rendimiento del portafolio de inversiones ya existente al inicio de 2026;
- $\widehat{RI}^{cetes}_{2026,s}$ es el rendimiento incremental de las nuevas reservas creadas durante 2026 e invertidas en CETES;
- $\widehat{RIF}^{res}_{2026,s}$ es el residual del resto de subcuentas del RIF.

### Resultado de inversiones base

Para estimar el rendimiento del portafolio ya existente al arranque de 2026, se utiliza el rubro "Resultado de inversiones" del estado de resultados y la cuenta "Inversiones" del balance general. Como el resultado anual es un flujo y las inversiones del balance son un saldo de cierre, se aproxima la base invertida de cada ano mediante inversiones promedio:

$$
InvProm_t = \frac{Inv_{t-1} + Inv_t}{2}, \qquad t \in \{2022,2023,2024,2025\}
$$

Despues se calcula la tasa anual de rendimiento base:

$$
y_t = \frac{RI_t}{InvProm_t}
$$

**[REVISA]** En esta seccion no se usa CAGR como estadistico principal porque $y_t$ ya es una tasa de rendimiento anual; aqui es mas natural comparar promedio anual, promedio ponderado por inversiones promedio y promedio reciente.

Una vez calculados los cuatro valores de $y_t$, se recomienda comparar al menos los siguientes estadisticos:

$$
\bar y = \frac{1}{4}\sum_{t=2022}^{2025} y_t, \qquad \bar y_w = \frac{\sum_{t=2022}^{2025} RI_t}{\sum_{t=2022}^{2025} InvProm_t}, \qquad \bar y_{rec} = \frac{y_{2024} + y_{2025}}{2}
$$

La proyeccion base del resultado de inversiones se formula como:

$$
\widehat{RI}^{base}_{2026,s} = \widehat y_s\,Inv_{2025}
$$

donde $Inv_{2025}$ se usa como aproximacion a las inversiones de apertura de 2026. Este componente representa el rendimiento del stock de inversiones heredado del cierre de 2025 y, por construccion, no incorpora todavia las nuevas reservas de 2026.

### Resultado de inversiones por CETES de reservas nuevas

El segundo componente se calcula con el esquema mensual solicitado. Si $Res_m$ representa el flujo neto mensual que efectivamente se manda a reservas y $r_m$ la tasa efectiva mensual asociada a CETES a 28 dias, entonces la nueva inversion del mes es:

$$
I_m = 0.5\,Res_m
$$

Si $B_m$ denota el saldo acumulado invertido en CETES al cierre del mes $m$, una forma natural de modelar la acumulacion es:

$$
B_m = (B_{m-1} + I_m)(1 + r_m)
$$

por lo que el ingreso financiero mensual asociado a este bloque puede expresarse como:

$$
RI^{cetes}_m = (B_{m-1} + I_m)r_m
$$

y el resultado anual incremental queda dado por:

$$
\widehat{RI}^{cetes}_{2026,s} = \sum_{m=1}^{12} RI^{cetes}_m
$$

### Residual del resto de subcuentas del RIF

El resto del RIF incluye subcuentas como valuacion, venta de inversiones, recargo sobre primas, resultado cambiario y otros conceptos que pueden tener drivers economicos distintos. Para no sobrecargar el modelo con supuestos poco defendibles, en esta version del anexo se adopta la simplificacion conservadora:

$$
\widehat{RIF}^{res}_{2026,s} = 0
$$

## Utilidad antes de impuestos, ISR y utilidad neta

La utilidad antes de ISR de cada escenario se obtiene como:

$$
\widehat{UAI}_{2026,s} = \widehat{UOp}_{2026,s} + \widehat{RIF}_{2026,s}
$$

Para la provision de ISR, la opcion mas limpia con informacion anual consolidada es usar una tasa efectiva historica:

$$
\tau_t = \frac{ISR_t}{UAI_t}
$$

y luego proyectar una tasa efectiva central $\widehat \tau$ mediante promedio anual, promedio reciente o el criterio que mejor represente la estabilidad historica. La provision 2026 se obtiene como:

$$
\widehat{ISR}_{2026,s} = \widehat \tau\,\widehat{UAI}_{2026,s}
$$

De esta manera, la utilidad neta proyectada queda dada por:

$$
\widehat{UN}_{2026,s} = \widehat{UAI}_{2026,s} - \widehat{ISR}_{2026,s}
$$

**[REVISA]** Si en la hoja final usaste directamente la tasa legal de ISR en vez de una tasa efectiva historica, sustituye esta subseccion para que el anexo refleje exactamente el calculo realizado.

## Supuestos y limites del modelo

- El modelo se construye con estados financieros consolidados anuales de 2021 a 2025; por tanto, no observa distribucion mensual real de ventas, vencimientos o siniestros.
- La proyeccion de 2026 es un ejercicio por escenarios y no una valuacion actuarial a nivel poliza.
- La metodologia privilegia relaciones economicas defendibles y consistencia contable sobre aparente precision estadistica.
- La descomposicion del RIF evita doble conteo entre el portafolio heredado de 2025 y la inversion en CETES de las nuevas reservas de 2026.
- El residual del RIF se fija en cero por prudencia metodologica, salvo que la revision final muestre que alguna subcuenta adicional es material y recurrente.

Con estas salvedades, la proyeccion 2026 queda sustentada por una cadena de supuestos explicitos, formulas verificables y relaciones contables consistentes con la estructura del estado de resultados.
