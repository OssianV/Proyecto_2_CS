# Explicacion del funcionamiento de la GUI

En esta libreta documentare el funcionamiento de la GUI. Esto servira para el manual de usuario.

> **Nota**: Por el momento la usare mas para organizar y planificar como va a quedar la GUI. 

## Modulo gui.py

Se creara un modulo para la gui. Este modulo tendra clases y funciones que se ejecutaran en el script principal.

### Librerias utilizadas

Para el desarrollo de la GUI se utilizaran las siguientes librerias:

- Tkinter: Implementacion de la GUI.
- logging: Rastreo de eventos (debugging y monitoreo).

## Estructura general de las ventanas

Cada ventana (a excepcion de la de error) tendra un boton para ir hacia atras o hacia delante (a menos que no haya un "atras" o "adelante"). Ademas, cada ventana tendra la opcion de minimizarze, extenderse, o cerrarse. Por ultimo, cada ventana tendra un titulo.

## Mensajes de error y confirmacion

Cuando suceda un error, se mandara generara una nueva ventana con un mensaje de error indicando el error. 

A su vez, cuando se suban los archivos .csv, se mostrara un mensaje que confirme que este proceso fue exitoso.
