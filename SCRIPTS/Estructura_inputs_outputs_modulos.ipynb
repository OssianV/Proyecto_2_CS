# Estructura basica de los inputs/outputs de cada script

---

En esta libreta de ipython se define la estructura basica de los inputs y putputs de cada script que desarrollaremos. Cada quien hara un script enfocado en areas especificas del proyecto. Dado que estos estan interrelacionados, es probable que sin la organizacion adecuada nuestros scripts no sean compatibles. Una opcion para arreglar esto es trabajar linealmente; esto es, que primero hagamos el script de la GUI, despues el de lectura de datos, despues el de calculo de razones, y por ultimo el de la generacion del dashboard. Si hicieramos esto, nadie podria avanzar hasta que el anterior hubiere terminado su parte. Es por esto que la mejor opcion es definir la estructura basica de los inputs y outputs de cada script, de esta manera quien haga el script de calculo de razones, pueda hacerlo sin ya tener el codigo de lectura de datos finalizado.

### Estructura de la libreta

Para cada script hay que definir dos cosas: (1) La estructura basica de sus **inputs**; y (2) la estructura basica de sus **outputs**. En (1) se definira la forma o estctura de aquellos objetos/datos que el script necesitara para poder desarrollarse. Un ejemplo seria la forma en la que se ingresaran las cuentas y sus valores para el desarrollo del modulo de calculo de razones financieras, el cual depende de dicha informacion (la cual se obtiene en el modulo de lectura de datos). En (2) se define la estructura que deben de tener algunos outputs de las funciones del modulo. Usando el mismo ejemplo de antes, el modulo de lectura de datos tendra que tener el formato que se definio tendria el input del modulo de calculo de razones financieras.

## Script GUI

El script de la GUI sera el script principal (donde se ejecutara el proyecto). En este se importaran los demas modulos y se usaran las funciones que estos contengan.

#### Estructura **INPUTS**

Por el momento, la GUI dependera de dos cosas:

- Resultado de la validacion de los archivos .csv realizada con funciones del modulo de lectura, limpieza y validacion.
- Los valores de las razones financieras, las cuales se desplegaran para la visualizacion de resultados.

El resultado de los archivos .csv sera un diccionario con dos pares llaves-valor. La estructura de cada par llave-valor sera "NOMBRE_ESTADO_FINANCIERO":(Bool, string_de_error). En el lugar de "NOMBRE_ESTADO_FINANCIERO" ira el nombre del estado financiero asociado a cada archivo. El valor (Bool, string_error) sera una tupla cuya primera entrada sera un booleano; si es TRUE entonces el archivo paso la validacion, si es FALSE no la paso. Si no pasa la validacion, entonces el string_error contendra un string especificando cuales fueron las validaciones que no paso el archivo.

La estructura de las razones financieras sera un diccionario de la forma:

`{"alias_razon": 0.34,
  "alias_razon": 0.76,
  ...
  "alias_razon": valor,
  ...
  }`

#### Estructura **OUTPUTS**

Por el momento, el unico output relevante de este script que puede afectar la funcionalidad de otros, son los *PATH* exactos de cada archivo .csv que el usuario ingrese. El modulo de lectura de datos hara uso de esto. Es por eso que se define que cada archivo estara asociado a una variable tipo string, las cuales tendran los *PATH* exactos antes mecionados.

## Modulo Lectura, limpieza y validacion de datos

El modulo de Lectura, limpieza y validacion de datos hara exactamente lo que su nombre describe: leer los datos de los .csv que el usuario proporciona, limpiar dichos datos y validar que la informacion que contengan los estados financieros tenga sentido.

#### Estrucura **INPUTS**

Los unicos inputs (por el momento) de este modulo son las direcciones absolutas (*absolute PATHs*) de los archivos .csv. La estructura de dicha info es basica. Habra dos variables tupo string, donde cada una contendra la direccion absoluta de uno de los dos estados financieros.

#### Estructura **OUTPUTS**

Por el momento, las funciones de este modulo tendran dos outputs principales:

- El diccionario con los nombres de la cuentas y sus valores.
- El diccionario que contendra el resultado de la validacion de cada estado fincanciero (si se consideraron validos o no)

La estructura del diccionario con los nombres de las cuentas y sus valores sera simple. Debera retornar un diccionarioc cuyos pares llave-valor sean "nombre_cuenta"-"valor". Para que no haya problemas lo ideal seria estandarizar los nombres de las cuentas, eso se los dejo a ustedes. Si alguien piensa en estandarizarlos, aqui escriban el formato de como deben quedan. 

Para el diccionario que contendra el resultado de la validacion de cada estado financiero tenemos que el output asociado sera un diccionario con dos pares llaves-valor. La estructura de cada par llave-valor sera "NOMBRE_ESTADO_FINANCIERO":(Bool, string_de_error). En el lugar de "NOMBRE_ESTADO_FINANCIERO" ira el nombre del estado financiero asociado a cada archivo. El valor (Bool, string_error) sera una tupla cuya primera entrada sera un booleano; si es TRUE entonces el archivo paso la validacion, si es FALSE no la paso. Si no pasa la validacion, entonces el string_error contendra un string especificando cuales fueron las validaciones que no paso el archivo; si la paso el string puede ser un string vacio.

## Modulo de calculo de razones financieras

Este modulo, como su nombre lo indica, sera en donde se realizaran los calculos para cada una de las razones financieras a analizar. Como extra, si alcanzamos a implementar las intepretaciones automaticas. En este modulo tambien se incluiran funciones que se encargen de eso.

#### Estructura **INPUTS**

El unico input que tendra (por el momento) este modulo es el diccionario de pares nombre_cuenta:valor, el cual se obtendra con las funciones del modulo de lectura de datos. Para que no haya problemas lo ideal seria estandarizar los nombres de las cuentas, eso se los dejo a ustedes. Si alguien piensa en estandarizarlos, aqui escriban el formato de como deben quedan.

#### Estructura **OUTPUTS**

Por el momento, el unico output de las funciones de este modulo, deberia ser un diccionario de la forma:

`{"alias_razon": 0.34,
  "alias_razon": 0.76,
  ...
  "alias_razon": valor,
  ...
  }`
