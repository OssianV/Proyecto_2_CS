# Formato basico de los archivos .csv de ingesta

Por el momento, nuestra aplicacion requerira la ingesta de dos archivos .csv. Un archivo estara asociado al Balance General y el otro al Estado de Resultados. En esta libreta se define el formato que deberan tener estos dos archivos.

> No pense tan a profundidad este formato. Si ven algo inconsistente o piensan que algo que defini solo complica las cosas, levanten un issue en github con su analisis del problema y si pueden planteen una solucion (o manden mensaje por wats, pero de preferencia un issue en github).

---
---

## Estructura del Balance General 

---
---

Ahora detallamos la estructura o formato del .csv asociado al balance general. Para no complicarnoslo, el formato es parecido al usado en clase, especificamente el usado en el archivo de excel donde realizamos el analisis de razones financieras. Para ver un .csv con el formato adecuado, vease el [.csv del Balance General](../INPUTS/Balance_general.csv).

### Posicionamiento de las cuentas

---

Para entender el formato de posicionamiento de las cuentas, tenemos que divirlo en partes:

- Estructura A = P+C
- Uso de columnas
- Uso de filas

#### Estructura A | P+C

Notese que la estructura del balance general sigue el formato de tener a la izquierda todas la cuentas de activos, y a la derecha las cuentas de pasivos, y abajo de estas, las del capital. Entendiendo esto, sera facil entender la repeticion en el formato que debe tener el archivo. Mas al respecto se definira mas adelante.

#### Uso de columnas

Para ver con mayor claridad la estructura de las columnas, vease la siguiente captura con las filas numeradas y con una descripcion del tipo de dato que se contendra en estas

![imagen](../ATTACHMENTS/_png/imagen_manual_de_usuario_002.jpg)

Como se puede ver, para cada seccion (activos y pasivos+capital), se cuentan con 6 columnas:

1. **Cuenta nivel 1**: En estas columnas (una en activos y otra en pasivos+capital), se pondran los nombres de las cuentas mas importantes. La nomenclatura "nivel n" que se usa en esta y en otras columnas es para senalar el nivel de generalidad de la cuenta. En el caso de esta columna, solo se incluiran las cuentas "Activo", "Pasivo" y "Capital", y sus asociadas "Suma del Activo", "Suma del Pasivo", "Suma del Capital" y "Suma del Pasivo y Capital".
2. **Cuenta nivel 2**: En esta columna se ponen cuentas de 2do nivel. Ejemplos son "Inversiones", "Inversiones para Obligaciones Laborales", "Disponibilidad", "Reservas tecnicas", "Acreedores", etc. Todas las cuentas en este nivel constituiran subcuentas de la inmediata cuanta principal anterior ("Activo", "Pasivo" o "Capital"). 
3. **Cuenta nivel 3**: De manera similar que con las cuentas nivel 2, las cuentas de este nivel seran a su vez subcuentas de la inmediata cuenta de nivel 2 anterior. Por ejemplo, en la cuenta nivel 1 podemos tener "Pasivos", en la cuenta nivel 2 "Reservas tecnicas" y en la cuenta nivel 3 "De Riesgos en curso".
4. **Cuenta nivel 4**: Mismo argumento. Siguiendo el ejemplo anterior, la cuenta nivel 4 podria ser "Seguros de Daños", refiriendose al pasivo de la reserva de riesgos en curso asociada a los seguros de danos.
5. **Valor cuenta nivel 3 y 4**: En esta columna es donde se pondra el valor total de las subcuentas del nivel 3 y 4. Simple.
6. **Valor cuenta nivel 1 y 2**: En esta columna es donde se pondra el valor total de las cuentas del nivel 1 y las subcuentas del nivel 2. Notese que no todas las cuentas del nivel 1 tendran valor asociado, solo las del tipo "Suma del ...".

Notese que este patron se repite tanto para la seccion asociada a los activos y a la asociada a los pasivos+capital. Esto siguiendo la estructura de separar en dos el balance general siguiendo la ecuacion contable (A = P + C).

Idealmente se deberia implementar dinamismo en el codigo para ajustarse a formatos mas laxos. Por ejemplo, que el codigo aun funcione incluso si entre la columna "Valor cuenta nivel 1 y 2" asociada a los activos y la columna " Cuenta nivel 1" asociadada a los pasivos+capital, hubiera varias columnas vacias (el codigo entonces limpiaria los datos y por consecuencia eliminaria dichas columnas).

#### Uso de las filas

La estrucura del uso de las filas es mas laxa. Las funciones de lectura dentro del modulo deben poder eliminar filas vacias a discrecion. Lo que si hay que tomar en cuenta es que en cada fila con datos significativos, solo se podra poner informacion de una cuenta/subcuenta.

### Nombre de las cuentas

---

Por el momento, los nombres de las cuentas seran exactamente los mismos que se encuentran en el [.csv](../INPUTS/Balance_general.csv). 

Lo unico a destacar, es que si el valor de una cuenta debe ser negativo (i.e. disminuir el valor de la cuenta de la cual es subcuenta), su nombre iniciara con el string "(-) " (parentesis_izquierdo - guion - parentesis_derecho - 1 espacio en blanco).

> **Nota**: Si se les occure que seria mejor usar los numeros asociados a las cuentas, o adaptar los nombres de acuerdo a alguna catalogo de cuentas, piensen bien como plantearlo y abran un issue en github (idealmente, sino tmb pueden mandar un wats) donde pongan su plan para solucionarlo. No empiecen a implementarlo hasta que los demas demos por valido dicho cambio (resolver el issue), solo planteen como hacerlo.

### Formato de valores numericos

---

Todos los valores seran positivos, del tipo flotante y usaran comas como separador de millares. Para las cuentas cuyo valor disminuya el valor de la cuenta de la cual son subcuenta, el valor sera positivo; para idenfitifiocar que dicho valor debe de restar, el nombre de la cuenta asociada iniciara con el string "(-) " (para mas lee la seccion "Nombre de las cuentas").

## Estructura del Estado de Resultados

---
---

Ahora detallamos la estructura o formato del .csv asociado al balance general. Igual que con el balance general, lo que aqui hacemos es simplemente describir el formato del estado de resultados en el archivo usado en clase (pues ese sera nuestro estandar). Para ver un .csv con el formato adecuado, vease este [.csv](..//INPUTS/Estado_de_resultados.csv).

### Posicionamiento de las cuentas

---

#### Uso de columnas

Para entender la estructura de las columnas de las cuentas vease la siguiente imagen:

![imagen](../ATTACHMENTS/_png/imagen_manual_de_usuario_006.jpg)

Notese que la 3er fila se agrego unicamente para facilitar la explicacion del formato, esta no forma parte de la estructura del .csv. Como se puede ver, se tienen 5 columnas:

1. **"Cuenta nivel 1"**: En esta columna se incluye el nombre de cuentas tales que las inmediatas cuentas posteriores en la columna "Cuenta nivel 2" consitituyan sus subcuentas.
2. **"Cuenta nivel 2"**: Esta columna incluye el nombre de subcuentas o cuentas de agregados. Con cuentas de agregados, me refiero a cuentas como "Utilidad Bruta", "Utilidad Neta", entre otras, i.e. cuentas que se definan como la suma de otras.
3. **"Valor jerarquia 1"**: En las columnas de valores, simplemente se ordenan los valores de las cuentas a las que estan asociadas. Si el valor se encuentra en la columna *Valor jerarquia 1*, dicho valor no puede depender de otras cuentas cuyos valores esten en las columnas *Valor jerarquia 2* o *Valor jerarquia 3*. Lo mismo aplica para las demas; si esta en *Valor jerarquia 2* solo puede depender de cuentas superiores en la misma o anterior columna; si esta en la 3ra columna, la explicacion es la misma.
4. **"Valor jerarquia 2"**: Vease explicacion de *Valor jerarquia 1*.
5. **"Valor jerarquia 3"**: Vease explicacion de *Valor jerarquia 1*.

#### Uso de filas

El uso de filas es basico. Una cuenta por fila, se permiten filas vacias (y por tanto, el modulo de lectura debe ajustarse a la posibilidad de filas vacias entre filas no vacias).

### Nombre de las cuentas y distincion entre cuenta de ingreso y egreso

---

Por el momento, los nombres de las cuentas seran exactamente los mismos que se encuentran en el [.csv](../INPUTS/Balance_general.csv). 

De la misma manera que con las cuentas del balance general, las cuentas que representen un gasto deberan iniciar con el string "(-1) ". Aunque en este caso, este estandar tiene un nivel de complejidad mas alto, pues en cierta manera se pueden *anidar* los (-1). Para entender esto vease la siguiente captura del [.csv con formato del Estado de Resultados](../INPUTS/Estado_de_resultados.csv):

![imagen](..//ATTACHMENTS/_png/imagen_manual_de_usuario_004.jpg)

Para entender el ejemplo de la imagen, observa que la cuenta principal *( -1 ) Costo Neto de Adquisicion* es un **gasto** debido a que tiene el prefijo "( -1 ) ". Despues subcuentas como *Comisiones a Agentes* representan entonces un **gasto**. En el ejemplo, la subcuenta *( - ) Comisiones por Reaseguro y Reafianzamiento Cedido* representa entonces un ingreso (piensa que el -1 se esta cancelando).

De esta manera, si una cuenta del nivel 1 no tiene el prefijo "( -1 ) ", entonces es de ingreso; y por tanto sus subcuentas sin dicho prefiijo seran de ingreso, y aquellas con el prefijo seran de gasto. Por otro lado, si una cuenta del nivel 1 tiene el prefijo, entonces es de egreso; y por tanto sus subcuentas sin el prefijo seran de gasto, y aquellas con el prefiijo seran de ingreso. **ENTIENDAN BIEN ESTO, QUE PUEDE CONFUNDIR**

### Formato de valores numericos

---

La mayoria de las cuentas tendran un valor positivo. Para determinar si este agrega o quita, sera necesario checar el tipo de cuenta (ingreso o egreso). Ademas, el valor puede ser negativo. Un valor negativo de una cuenta de gasto representa un ingreso o una reversion de un gasto anterior. Por lo demas, el valor sera flotante y se usaran comas para separar los millares.

Para ver un ejemplo del valor de una cuenta de egreso negativa (lo cual implica un ingreso), vean la siguiente imagen:

![imagen](../ATTACHMENTS/_png/imagen_manual_de_usuario_005.jpg)

Como se puede ver, la cuenta de nivel 1 *Resultado Integral de Financiamiento* es una cuenta de ingreso (pues no tiene el prefijo "( -1 ) "), por lo tanto la cuenta *( - ) Castigos preventivos por importes recuperables de reaseguro* es una cuenta de gasto (pues si tiene el prefijo). Como se ve en la imagen, el valor asociado a esta cuenta es `-25,442.87`, por lo tanto esta cuenta de gasto representara un ingreso (mas especificamente, una reversión de un gasto anterior, pues esta cuenta en concreto es una partida *estimativa*).
