```python
import pandas as pd
import re
```

```python
# Función para extraer número

def extraer_numero(valor):
    if pd.isna(valor):
        return None
    
    match = re.search(r'-?\d+[\d,]*\.?\d*', str(valor))
    if match:
        return float(match.group().replace(",", ""))
    return None
```

```python
def construir_diccionario(path_csv, cuentas_objetivo):
    df = pd.read_csv(path_csv, header=None)
    
    resultado = {}
    
    for texto_csv, llave_dict in cuentas_objetivo.items():
        encontrado = False
        
        for _, row in df.iterrows():
            for j, celda in enumerate(row):
                
                if str(celda).strip().lower() == texto_csv.lower():
                    
                    for k in range(j + 1, len(row)):
                        numero = extraer_numero(row[k])
                        if numero is not None:
                            resultado[llave_dict] = numero
                            encontrado = True
                            break
                    break
            
            if encontrado:
                break
        
        if not encontrado:
            resultado[llave_dict] = None
    
    return resultado
```
