## Proyeccion de ingresos

En esta seccion se cubre todo lo pertienente a los ingresos que recibe la aseguradora sin tomar en cuenta avaluo de inversiones.

Notese que nos centramos primero en proyectar el total asociado al ciclo 2026. Solo despues pensaremos en como amortizar esto a lo largo de los meses de 2026.

### Prima emitida

El profesor recomendo que para realizar la proyeccion de la prima emitida del 2026, utilizaramos el promedio interanual de las tasas de incremento/decremento de la prima emitida en años consecutivos.

El problema con esa proyeccion es que toma en cuenta escenarios *anormales* asociados al crecimiento de los años 2022 y 2023. Una googleada rapida confirmo que el sector vida en Mexico tuvo un incremento anormal posterior a la pandemia COVID-19 (falta investigar bien). Esto podria indicar que ese tipo de crecimiento no es de esperarse(pues no podemos esperar otra pandemia). Por lo cual, el dar el mismo peso a las tasas de incremento del 2022 y 2023, o inclusive tomarlas en cuenta en su totalidad, no refleja la futura realidad.

- Tasas de incremento en primas cedidas: 2022: 11.59%	2023: 13.59%	2024: 6.49%	  2025: 7.09%

Para evaluar que tanto afecta esta diferencia, lo que plantee es lo siguiente:

#### Promedio interanual vs. CAGR vs. promedio ultimos dos años

Calcule tanto el promedio interanual (lo que dijo el profe), el CAGR (Coumpound Anual Growth Rate), y el promedio interanual solo tomando en cuenta los ultimos dos años. El resultado obtenido fue el siguiente:


- Promedio interanual: 9.69%
- CAGR 2021-2025: 7.65%
- Promedio ultimos dos años: 6.79%

Como se puede ver, hay una diferencia de casi 3 puntos porcentuales entre el promedio interanual que toma en cuenta los 4 periodos, y el promedio de los ultimos dos años. Esto, aunado al evidente pico en el crecimiento en los intervalos 2021-2022 y 2022-2023, sugiere que usar el promedio interanual no seria lo mas preciso.

El **CAGR** provee una tasa de crecimiento mas realista, pues aunque este si toma en cuenta los años anormales, no les da a todos el mismo peso, sino que se enfoca directamente en las primas emitidas del 2025 y del 2021.

#### Mi proposicion para el calculo

Lo mejor que se me ocurre es plantear escenarios. Especificamente 3:

- **Escenario de normalizacion post-pandemia**: Asumimos que el resultado de la pandemia fue transitorio, y que el crecimiento del 2026 debe parecerse mas al ritmo reciente. Usamos el promedio de los ultimos dos años.
- **Escenario de persistencia parcial del impulso post-pandemia**: Asumimos que el crecimiento del sector despues de la pandemia se desacelera, pero se sigue notando la afectacion del crecimiento anormal, pero en menor medida. Usamos el CAGR.
- **Escenario del profe**: Le damos el mismo peso a todos los años, anque no sea lo mas sensato. Usmaos el promedio interanual de todo el periodo.

- Promedio de propocion anual: 3.68%

Como ambas dan practicamente lo mismo, nos decidimeros por utilizar el promedio de proporcion anual.

### Primas retenidas devengadas

Las primas retenidas devengadas son el resultado de la diferencia:

$$\text{Primas retenidas} - \text{Incremento neto en la reserva de riesgos en curso y fianzas vigentes}$$

Tenemos dos opciones: (1) *Estimar el Incremento neto en la reserva de riesgos en curso y fianzas vigentes* y restarlo a la estimacion de las prims retenidas, o (2) directamente *estimar las primas retenidas devengadas*.

Antes hay que entender a que nos referimos con el incremento neto en la reserva de riesgos en curso y fianzas vigentes:

#### Incremento neto en la reserva de riesgos en curso y fianzas vigentes

Esta cuenta describe el cambio de todas las primas no devengadas con la que la aseguradora cuenta al final del ciclo contable, con respecto a aquellas con las que inicio. 

Si, por ejemplo, uno paga un seguro de vida anual en el 31 de octubre, hasta el 31 de diciembre (fecha de cierre contable), solo han transcurrido 2 meses desde el inico de la poliza. Entonces, el ingreso asociado a esos dos meses sera contabilizado como prima devengada, pero el asocioado a los 10 meses restantes, no. Entonces, el total de las primas asociado a esos 10 meses, debera estar incluido en el incremento neto de las reservas de riesgos en curso y fianzas en vigor.

Estas primas no devengadas se registran en la reserva de riesgos en curso. Lo que mide esta cuenta es el cambio neto durante el año en estas reservas. Escencialemte haciendo:

$$ \text{Incremento Neto}_t = \text{Reserva final}_t - \text{Reserva inicial}_t $$

#### Que vamos a estimar?

Lo mas sensato es estimar las primas retenidas devengadas. La cuenta de incremento neto en las reservas de riegos en curso no es mas que una cuenta de balance, no es nuestro interes principal. 

#### Como vamos a estimar las primas retenidas devengadas?

Para realizar la proyeccion, prestaremos atencion a la razon:

$$\text{Razon primas retenidas devengadas/primas retenidas}_i = \frac{\text{Primas retenidas devengadas}_i}{\text{Primas retenidas}_i}$$

En base a esas razones, calcularemos la tasa de devengamiento a utilizar para proyectar las primas retenidas devengadas. Se me ocurren tres maneras para calcular la tasa a utilizar:

1. Promedio interanual de todo el periodo: 94.57%
2. Promedio interanual de los ultimos dos anos: 96.48%
3. Promedio ponderado de todo el periodo: 94.68%

Notese que el promedio interanual de los ultimos dos anos es superior por casi dos puntos porcentuales a los otros dos estadisticos.

Esto probablemente se deba a el incremento anormal en la prima emitida en los periodos 2021-2022 y 2022-2023. La explicacion de esto es que como se emitieron muchas nuevas polizas, se incremento la reserva de riesgos en curso, cuyo incremento se termino restando a la prima retenida.

En base a los resultados obtenidos, el tipo de calculo a utilizar por escenario planteado sera:

- **Escenario de normalizacion post-pandemia**: Promedio interanual de los ultimos dos anos.
- **Escenario de persistencia parcial del impulso post-pandemia**: Promedio ponderado de todo el periodo
- **Escenario del profe**: Promedio ponderado del periodo

De esta manera, obtendremos la proyeccion de las primas retenidas devengadas haciendo:

$$\text{Primas Retenidas Devengadas}_{2026} = \text{Primas Retenidas}_{2026} \times \text{Tasa de devengamiento}$$

### Costo de adquisicion

El costo de adquisicion no es nada mas que el precio que pagamos por obtener los ingresos del ejercicio (no contando por operar durante este). Lo principal son las comisiones y compensaciones a los agentes de seguro; aunque tambien se puede incluir el marketing y otras gastos. Lo natural es entonces proyectar el costo de adquisicion del 2026, tomando como referencia la proporcion que historicamente ha significado este costo, con respecto a la totalidad de las primas emitidas. Dichas proporciones son:

- 2021:	0.139705664
- 2022:	0.140378117
- 2023:	0.172375256
- 2024:	0.151338406
- 2025:	0.149766818

El reto es ahora decidir que approach tomar para obtener el estadistico a utilizar para proyectar el costo de adquisicion del 2026. Para esto voy a plantear dos approaches: **(1)** promedio anual de las proporciones, y **(2)** promedio ponderado del periodo completo.

Voy a utilizar el modelar el costo de adquisicion como ejemplo para mostrar la diferencia entre el promedio anual y el promedio ponderado.

#### Promedio de proporcion anual

Calcular el promedio de proporcion anual es bastante basico. Simplemente hacemos:

$$\overline{\text{Proporcion anual}} = \frac{\sum_{k=2021}^{2025}\text{Proporcion}_k}{5}$$
 
Este promedio da el mismo peso a cada una de las observaciones, y por ello es suceptible a outliers. No es idoneo el utilizarlo en algunas circumstancias, pues puede osbreestimar la tendencia de largo plazo cuando hay mucha volatilidad.

#### Promedio ponderado de proporcion anual

Calcular este promedio es tambien bastante basico:

$$\text{Promedio ponderado} = \frac{\sum_{t=2021}^{2025}C_t}{\sum_{t=2021}^{2025}P_t} = \frac{\sum_{t=2021}^{2025}P_t \times r_t}{\sum_{t=2021}^{2025}P_t}$$

donde $P_t$ es la prima emitida en el ano t; $C_t$ es el costo neto de adquisicion en el ano t, y $r_t$ es la proporcion del costo con respecto a la prima emitida del ano t.

Este promedio le da mas peso a los anos donde haya habido mayor prima emitida, por lo cual es menos suceptible a outliers, pues les da un mayor o menor peso (dependiendo de si fue una mayor o menor prima).

#### Seleccion del modelo

Calculando ambos estadisticos obtenemos:

- Promedio interanual: 15.07%
- Promedio ponderado del periodo: 15.14%

Como ambas dan casi lo mismo, usaremos el **promedio ponderado del periodo**, el cual es mas robusto. 

#### Promedio ponderado de la proporcion marginal
> A este no le hagan caso.

El promedio ponderado del periodo es similar. Para calcularlo hacemos:

$$ \overline{\text{Porporcion anual}}_w = \frac{\text{Costo Neto de Adq.}_{2025} - \text{Costo Neto de Adq.}_{2021}}{\text{Primas emitidas}_{2025} - \text{Primas emitidas}_{2021}} $$

Para entender porque este es un promedio pondedaro, observese que:

$$\frac{\text{Costo Neto de Adq.}_{2025} - \text{Costo Neto de Adq.}_{2021}}{\text{Primas emitidas}_{2025} - \text{Primas emitidas}_{2021}} = \frac{\sum_{k=2022}^{2025}\vartriangle \text{Costo neto de Adq.}_{k}}{\sum_{k=2022}^{2025}\vartriangle \text{Primas emitidas}_{k}}$$

Pero la proporcion marginal anual del costo neto de adquisicion con respecto a las primas emitidas es:

$$P_t = \frac{\vartriangle \text{Costo neto de Adq.}_t}{\vartriangle \text{Primas emitidas}_t}$$

> Esta proporcion marginal representa la proporcion del incremento/decremento del costo neto de Adq. con respecto al cambio total en las primas emitidas en el ciclo (t-1) - (t).

Despejando el cambio de las primas emitidas obtenemos que

$$ \vartriangle \text{Costo neto de Adq.}_t = P_t \times \vartriangle \text{Primas emitidas}_t$$

Por lo tanto:

$$\overline{\text{Porporcion anual}}_w = \frac{\sum_{k=2022}^{2025}(P_k \times \vartriangle \text{Primas emitidas}_k)}{\sum_{k=2022}^{2025}\vartriangle \text{Primas emitidas}_{k}}$$

Donde la proporcion anual es el promedio ponderado de la proporcion anual; y donde el conjunto `W` de pesos es: 

$$W=[w_i]_{2022 \leq i \leq 2025}=[\vartriangle \text{Primas emitidas}_{i}]_{2022 \leq i \leq 2025}$$

Por lo tanto, el peso que se le da a cada periodo depende de que tanto incremento hubo en las primas emitidasl; donde no se esta promediando la proporcion del costo de adq. con respecto a dichas primas, sino la **proporcion marginal del costo de adq.** con respecto a estas.

### Costo neto de siniestralidad

Esta es una parte importante donde cambia el aproach al escogido por el profe.

Aqui vale la pena preguntarse, que tiene mas sentido? Decir que la siniestralidad *crece* a un cierto porcentaje fijo, o decir que por cada peso de prima retenida devengada se espera cierto costo de siniestros? A mi parecer, claramente la respuesta es la ultima.

Y tiene sentido. El costo neto de siniestralidad tiende a ser mas volatil que las primas, y este es propenso a eventos atipicos (como fue el COVID-19). Es por esto que es mas sensato calcular la razon historica:

$$\text{razon de perdida} = \frac{\text{Costo neto de siniestralidad}}{\text{Primas Retenidas Devengadas}}$$

#### Por que las primas retenidas devengadas?

Tenemos realmente tres opciones bajo las cuales podriamos calcular la razon: **(1)** Primas emitidas; **(2)** primas retenidas, y **(3)** primas retenidas devengadas. Bajo que criterio decidimos escoger las **primas retenidas devengadas**?

La idea es que el costo de siniestralidad es una cuenta de resultado tecnico, y por tanto, lo mas sensato es pensar sobre que base economica se genera? Claramente no es sobre primas emitidas, pues una parte de esta se cede a reaseguradoras que se hacen cargo de cubrir su proporcion de pago de siniestros. Entonces la desicion esta entre las primas retenidas y retenidas devengadas.

Esta desicion es mas sutil. La idea por la cual seleccionamos las primas retenidas devengadas es que estas reflejan los ingresos del periodo contable actual. Hacer uso de las primas retenidas no es tan preciso, pues toma en cuenta primas re riesgo que todavia no han corrido. Eso tiene a subestimar la razon de perdida, pues se contabilizan las primas de riesgos que no son vigentes (no devengados). Estas primas se contabilizaran como ingreso el ciclo siguiente, contaminando la informacion.

#### Calculo de la proporcion a utilizar

La manera en la que decidamos proyectar el costo de siniestralidad debe hacer match con los escenarios propuestos en el calculo de la prima emitida. Primero revisamos la estabilidad de la razon calculando lo siguiente:

- Promedio anual	70.08%
- Promedio anual ultimos dos anos	71.00%
- Promedio ponderado del periodo	76.65%

Como se puede ver, la unica diferencia significativa es en el calculo del promedio ponderado del periodo. Como la diferencia entre el promedio anual y el de los utlimos dos anos es infima, decidimos usar el promedio anual para todos los escenarios.

### Resultado Integral del Financiamiento

La unica cuenta que vale la pena modelar en este ambito es el Resultado Integral del Financiamiento. El unico detalle a tomar en cuenta es el siguiente:

El profesor nos indico que las primas que se vayan recibiendo en 2026 de manera mensual (en la proyeccion), una vez restadas los gastos, se pasan a reservas las cuales se deben invertir en cetes a 28 dias. 

Si planeamos proyectar el resultado integral del financiamiento utilizando la informacion historica, tenemos que tener cuidado con contabilizar dos veces los rendimientos obtenidas de la inversion de las primas que se obtienen en 2026. Una opcion seria el no proyectar la inversion mensual de las reservas, pero eso mataria el proposito de la proyeccion. La otra es encontrar una manera de proyectar solo aquella parte de la RIF no asociada con el avaluo de la reserva que se obtiene a lo largo del 2026, y al final sumar la suma de los intereses del 2026 con ese total para dar el total del RIF.

A continuaiocn detallo mi desicion de como realizar el modelo:

#### Dividimos en tres el RIF

La manera en que yo planteo el modelar el **RIF** es el divirlo en tres:

1. Resultado de Inversiones sin contabilizar inversiones de primas recibidas en el 2026
2. Resultado de Inversiones de primas recibidas en el 2026
3. Residual del resto de componentes ajustado por escenario.

De esta manera, el RIF proyectado para el 2026 debe estar dato como:

$$\text{RIF}_{2026} = \text{RIF base}_{2026} + \text{RIF reservas credas en 2026} + \text{RIF residual}$$

#### Resultado de Inversiones sin contabilizar inversiones de primas recibidas en el 2026

Esta seccion cubre el resutlado del 2026 de las inversiones que la aseguradora tenia al finalizar el ciclo contable del 2025, sin contar las inversiones asociadas a las primas que se reciben en 2026.

La idea es la siguiente:

Calcular las inversiones promedio de 2022 a 2025. Esto se hace haciendo:

$$\text{Inversiones promedio}_{t} = \frac{\text{Inversiones al cierre de t-1} + \text{Inversiones al cierre de t}}{2}$$

La cuenta de inversiones al cierre se extrae de los balances generales, de la cuenta de activo "Inversiones".

Despues, para cada uno de los promedios, calculamos la tasa:

$$y_t = \frac{\text{Resultado de inversiones}_t}{\text{Inversiones promedio}_t}$$
